```markdown
# Approach 1

We will use a recursive function because we can't use a loop. This function takes an integer and an array as an argument. The base case for the function is when the 'argument integer' becomes zero. Otherwise, we will call the recursive function 'current element - 1' and insert the current element in the array.

The steps are as follows:

```java
// Recursive Function to push elements
function recursiveFunction(int x, int[] arr):
    If 'x == 0':
        Return;
    Call recursiveFunction(x - 1, arr).
    Insert 'x' into the array 'arr'.

// Function to get an array from 'n' to '1' without loop
function printNos(int x):
    Create an empty integer array 'arr'.
    Call recursiveFunction(x, arr).
    Return 'arr'.
```

### Time Complexity
O(n), where n is the integer.

We will call the recursive function 'n' times.

Hence, the time complexity is O(n).

### Space Complexity
O(n), where n is the integer.

'n' recursive functions will be called.

Hence, the space complexity is O(n).

### Java Solution

```java
/*
    Time Complexity: O(n)
    Space Complexity: O(n)

    where n is the integer.
*/
public class Solution {
    public static int[] printNos(int x) {
        // Declaring an empty integer array
        int[] ans = new int[x];

        // Calling recursive function
        recursiveFunction(x, ans);

        return ans;
    }

    public static void recursiveFunction(int x, int[] ans) {
        if (x == 0) {
            return;
        }

        // Call recursive function
        recursiveFunction(x - 1, ans);

        // Insert element in the array
        ans[x - 1] = x;
    }
}
```

### Explanation

1. The `printNos` function is the main function that takes the integer `x` as input and returns an array of integers from 1 to `x`.
2. It first declares an empty integer array `ans` of size `x` to store the result.
3. It then calls the `recursiveFunction` helper function, passing `x` and the `ans` array as arguments.
4. The `recursiveFunction` is a recursive helper function that performs the actual task of populating the `ans` array with integers from 1 to `x`.
5. The base case for the recursion is when `x` becomes 0, indicating that all elements have been added to the array.
6. In the recursive case, the function calls itself with `x - 1` as the argument, effectively counting down from `x` to 1.
7. After the recursive call, it inserts the current value of `x` into the `ans` array at the index `x - 1`.
8. The `printNos` function finally returns the populated `ans` array.
```



### Python
```python
from typing import List

def recursiveFunction(x: int, ans: List[int]) -> None:
    if x == 0:
        return
    # Call recursive function.
    recursiveFunction(x - 1, ans)
    # Insert element in the list.
    ans.append(x)

def printNos(x: int) -> List[int]:
    # Declaring an empty list.
    ans = []
    # Calling recursive function.
    recursiveFunction(x, ans)
    return ans
```

# Approach 2 space optimized
We will use a recursive function because we can’t use a loop. This function takes an integer and an array as an argument. The base case for the function is when the ‘argument integer’ becomes equal to ‘n’. Otherwise, we insert the current element in the array and call the recursive function ‘current element + 1’.

 

The steps are as follows:-
 

// Recursive Function to push elements

function recursiveFunction(int x, int[] arr , int n):

Insert ‘x’ into the array ‘arr’.
If ‘x == n’:
Return;
Else:
Return  recursiveFunction(x + 1, arr , n);
 
// Function to get an array from ‘n’ to ‘1’ without loop

function printNos(int x):

Create an empty integer array ‘arr’.
Call recursiveFunction(1, arr , n).
Return ‘arr’.
Time Complexity
O( n ), where n is the integer.
 

We will call the recursive function ‘n’ times.

 

Hence, the time complexity is O( n ).

Space Complexity
O( 1 ), 

 

Even though ‘n’ recursive functions will be called, only 1 will be present in the call stack at a time.
 

Hence, the space complexity is O( 1 ).


### JAVA 
```java
/*
    Time Complexity: O(n)
    Space Complexity: O(1)

    where n is the integer.
*/
import java.util.ArrayList;
import java.util.List;

public class Solution {
    public static void recursiveFunction(int x, List<Integer> ans, int n) {
        // Insert element in the list
        ans.add(x);
        if (x == n) {
            return;
        }

        // Call recursive function
        recursiveFunction(x + 1, ans, n);
    }

    public static List<Integer> printNos(int n) {
        // Declaring an empty list
        List<Integer> ans = new ArrayList<>();

        // Calling recursive function
        recursiveFunction(1, ans, n);

        return ans;
    }
}
```

### Python

```python
from typing import List

def recursiveFunction(x: int, ans: List[int], n: int) -> None:
    """
    Time Complexity: O(n)
    Space Complexity: O(1)

    where n is the integer.
    """
    # Insert element in the list
    ans.append(x)
    if x == n:
        return

    # Call recursive function
    recursiveFunction(x + 1, ans, n)

def printNos(x: int) -> List[int]:
    # Declaring an empty list
    ans = []

    # Calling recursive function
    recursiveFunction(1, ans, x)

    return ans
```
